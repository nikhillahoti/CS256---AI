```python
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.pyplot as plt
```

```python
# Data Creation
X, y = make_blobs(n_samples=100,n_features=2,centers=3, cluster_std=0.5,shuffle=True, random_state=0)

# Plot the graph
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.show()
```

```python
# plot for KMeans Inertia
n = 11
arr_X = []
arr_Y = []
for i in range(2, n):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X, y)
    arr_Y.append(kmeans.inertia_)
    arr_X.append(i)

# Plot for Inertia from K=2 to 10
plt.scatter(arr_X, arr_Y, marker='o', edgecolor='k')
plt.show()
```

```python
# The best value for k was obtained for k=3 because 
# because as the size of k increases the distance of the points from the centroids keep on decreasing
# but as per the elbow method, the point at which the inertia flattens, is the ideal size of the 
# number of clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_means = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_means, cmap='viridis')
plt.show()
```

```python
# Silhouette Score
n = 11
arr_X = []
arr_Y = []
for i in range(2, n):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X, y)
    y_pred = kmeans.predict(X)
    # The Silhouette score returns the average 
    arr_Y.append(silhouette_score(X, y_pred))
    arr_X.append(i)

# Plot for Silhouette Coefficient from K=2 to 10
plt.scatter(arr_X, arr_Y, marker='o', edgecolor='k')
plt.show()


# The Silhouette Coefficient is the highest for K=3.
# Hence the ideal value for k=3
```

```python
# DBScan Algorithm
dbScn = DBSCAN(eps=0.75, min_samples=3)
dbScn.fit(X)
y_means = dbScn.labels_
plt.scatter(X[:, 0], X[:, 1], c=y_means, cmap='viridis')
plt.show()
```

```python
# Make Moons
X, y = make_moons(n_samples=200,noise=0.05, random_state=0)

# Plot the graph
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.show()
```

```python
# DBScan for moons dataset
dbScn = DBSCAN(eps=0.2, min_samples=2)
dbScn.fit(X)
y_means = dbScn.labels_
plt.scatter(X[:, 0], X[:, 1], c=y_means, cmap='viridis')
plt.show()
```

```python
# KMeans for moons dataset
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_means = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_means, cmap='viridis')
plt.show()
```

```python
# The DBSCAN algorithm works the best for the moons data. 
# It is able to separate the clusters into two different classes as expected.
# K-Means is not able to achieve the results and it breaks the clusters
```
